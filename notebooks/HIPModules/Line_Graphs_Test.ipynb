```python
%%html

<script>
  function code_toggle() {
    if (code_shown){
      $('div.input').hide('500');
      $('#toggleButton').val('Show Code')
    } else {
      $('div.input').show('500');
      $('#toggleButton').val('Hide Code')
    }
    code_shown = !code_shown
  }

  $( document ).ready(function(){
    code_shown=false;
    $('div.input').hide()
  });
</script>
<form action="javascript:code_toggle()"><input type="submit" id="toggleButton" value="Show Code"></form>
```

```python
import plotly.graph_objs as go
import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
import plotly.tools as tls
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from IPython.display import HTML
import random
%matplotlib inline
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt 
import IPython
from IPython import display 
from ipywidgets import widgets
from ipywidgets import interactive
import warnings
warnings.filterwarnings("ignore")
```

```python
init_notebook_mode(connected=True) #initiating

iplot([{"x": [0, 1, 2], "y": [3, 6, 7]}]) #points plotting


fig = plt.Figure()
ax = fig.gca()
x = np.linspace(0, 1, 10) #start, end, evenly spaced numbers
y = abs(x) #makes all array values positive

#need to figure out what this stuff does
ax.plot(x,y) 
canvas = FigureCanvas(fig)
plotly_fig = tls.mpl_to_plotly(fig)

py.iplot(plotly_fig) #plots the thing

```

```python
N = 500
random_x = np.linspace(0, 1, N) #start, end, evenly spaced numbers
random_y = np.random.randn(N) #500 values

# Create a trace
trace = go.Scatter( #creates points from array
    x = random_x,
    y = random_y
)

data = [trace]

py.iplot(data, filename='basic-line')
```

```python
start_button = widgets.Button(description = "Update")
# val = widgets.FloatSlider(value = 0.5, min = 0.0, max = 1.0, description = "Multiplier")
val = 10

temp = np.linspace(-10,10,100) #start, stop, number of evenly spaced points
# print(temp)
my_x = np.array(temp)*val #multiplying every array item by val
# print(my_x)
my_y = [x*x*x*x*x*4 for x in my_x] #list comprehension
# print(my_y)

# Create a trace
trace = go.Scatter(
    x = my_x,
    y = my_y
)

data = [trace]

py.iplot(data, filename='basic')
```
